```python
import os
import openai
import numpy as np
from tqdm import tqdm
import pandas as pd

# Load the OPENAI KEY from the env.sh file
from dotenv import load_dotenv
load_dotenv('D:\Rishika-rai-Wassertoff-AITask\Rishika-rai-Wassertoff-AITask\env.sh')

def generate_cot_prompt(question, steps):
    """
    Generate a chain-of-thought prompt for a given question and a list of steps.

    :param question: The question to generate the prompt for.
    :param steps: A list of steps in the chain-of-thought process.
    :return: A string containing the prompt.
    """
    prompt = f"Q: {question}\n"
    for i, step in enumerate(steps, 1):
        prompt += f"{i}. {step}\n"
    prompt += "A: "
    return prompt

cot_steps = [
    "The Wasserstoff website is dedicated to using open technology to solve the world's biggest problems.",
    "The website believes that technology should be accessible to everyone, not just a select few.",
    "Considering these aspects, let's create a chain of thought for the website."
]

question = "What is the mission of the Wasserstoff website?"
prompt = generate_cot_prompt(question, cot_steps)
print(prompt)
```
